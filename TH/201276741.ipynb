# ĐỊNH NGHĨA - CHỨC NĂNG - ỨNG DỤNG

LSTM (Long Short-Term Memory):

Định nghĩa: LSTM là một loại mạng nơ-ron hồi quy (RNN) được thiết kế để giải quyết vấn đề phụ thuộc xa trong quá trình học của RNN thông thường. LSTM được giới thiệu bởi Sepp Hochreiter và Jürgen Schmidhuber vào năm 1997.

Chức năng: LSTM có các cổng (input, forget, output) và ô trạng thái bên trong, giúp mô hình học được các mối quan hệ phụ thuộc xa và giữ lại thông tin quan trọng trong chuỗi dài.
Ứng dụng: Dịch máy, nhận dạng giọng nói, phân tích cảm xúc trong văn bản, dự báo chuỗi thời gian, sinh văn bản tự động, và nhiều tác vụ xử lý ngôn ngữ tự nhiên khác.

BiLSTM (Bidirectional LSTM):

Định nghĩa: BiLSTM là một biến thể của LSTM, trong đó mô hình học cả hai hướng của chuỗi đầu vào - từ trái sang phải và từ phải sang trái.

Chức năng: BiLSTM giúp mô hình có thể học được nhiều thông tin hơn từ chuỗi đầu vào, bao gồm cả ngữ cảnh trước và sau một điểm dữ liệu cụ thể.

Ứng dụng: BiLSTM được sử dụng trong các ứng dụng tương tự như LSTM, nhưng có khả năng hiểu ngữ cảnh tốt hơn.

LSTM with Attention:

Định nghĩa: LSTM with Attention là một cải tiến của mô hình LSTM thông thường, bổ sung một cơ chế chú ý (attention) để giúp mô hình tập trung vào những phần quan trọng hơn của chuỗi đầu vào khi đưa ra dự đoán.

Chức năng: Cơ chế chú ý giúp mô hình xác định những phần nào của chuỗi đầu vào là quan trọng nhất để đưa ra quyết định cho từng bước thời gian trong chuỗi đầu ra.

Ứng dụng: LSTM with Attention được sử dụng rộng rãi trong các bài toán dịch máy tự động (seq2seq), tổng hợp tin tức, phân loại văn bản, và các tác vụ xử lý ngôn ngữ tự nhiên.

BiLSTM with Attention:

Định nghĩa: BiLSTM with Attention kết hợp cơ chế chú ý (attention) với mạng BiLSTM, giúp mô hình tận dụng lợi thế của cả hai hướng trong chuỗi đầu vào và đồng thời tập trung vào các phần quan trọng hơn trong chuỗi.

Chức năng: Với cơ chế chú ý, mô hình BiLSTM có thể xác định và tập trung vào những phần quan trọng nhất của chuỗi đầu vào từ cả hai hướng, giúp cải thiện hiệu suất của mô hình.

Ứng dụng: BiLSTM with Attention được sử dụng trong các ứng dụng tương tự như LSTM with Attention, bao gồm dịch máy tự động (seq2seq), tổng hợp tin tức, phân loại văn bản, và các tác vụ xử lý ngôn ngữ tự nhiên khác mà cần đến sự hiểu biết về ngữ cảnh và tập trung vào các phần quan trọng của chuỗi đầu vào.

Nhìn chung, LSTM, BiLSTM, LSTM with Attention và BiLSTM with Attention đều là những phương pháp hiệu quả để xử lý các bài toán liên quan đến chuỗi dữ liệu, đặc biệt là trong lĩnh vực xử lý ngôn ngữ tự nhiên. Tuy nhiên, các mô hình này có những đặc điểm và ưu nhược điểm riêng, do đó việc lựa chọn mô hình phù hợp sẽ phụ thuộc vào bài toán cụ thể và yêu cầu của ứng dụng.

## HÀM - LỚP

```python
class BahdanauAttention(tf.keras.layers.Layer):
    def __init__(self, units):
        super(BahdanauAttention, self).__init__()
        self.W1 = tf.keras.layers.Dense(units)
        self.W2 = tf.keras.layers.Dense(units)
        self.V = tf.keras.layers.Dense(1)

    def call(self, query, values):
        query_with_time_axis = tf.expand_dims(query, 1)
        values_with_time_axis = tf.expand_dims(values, 1)

        score = self.V(tf.nn.tanh(self.W1(values_with_time_axis) + self.W2(query_with_time_axis)))

        attention_weights = tf.nn.softmax(score, axis=1)

        context_vector = attention_weights * values_with_time_axis
        context_vector = tf.reduce_sum(context_vector, axis=1)

        return context_vector, attention_weights
```

Lớp Bahdanau Attention:

- class BahdanauAttention(tf.keras.layers.Layer): Khai báo một lớp BahdanauAttention kế thừa từ lớp Layer của TensorFlow.

- def __init__(self, units): Hàm khởi tạo của lớp, nhận vào tham số units - số lượng đơn vị trong các lớp Dense.

- super(BahdanauAttention, self).__init__(): Gọi hàm khởi tạo của lớp cha Layer.

- self.W1 = tf.keras.layers.Dense(units): Khai báo lớp Dense W1 với số lượng đơn vị là units.

- self.W2 = tf.keras.layers.Dense(units): Khai báo lớp Dense W2 với số lượng đơn vị là units.

- self.V = tf.keras.layers.Dense(1): Khai báo lớp Dense V với số lượng đơn vị là 1.

- def call(self, query, values): Hàm call của lớp, nhận vào hai tham số query và values.

- query_with_time_axis = tf.expand_dims(query, 1): Mở rộng kích thước của query theo chiều thứ 1, tạo thành query_with_time_axis.

- values_with_time_axis = tf.expand_dims(values, 1): Mở rộng kích thước của values theo chiều thứ 1, tạo thành values_with_time_axis.

- score = self.V(tf.nn.tanh(self.W1(values_with_time_axis) + self.W2(query_with_time_axis))): Tính toán giá trị score bằng cách áp dụng hàm tanh vào tổng của W1(values_with_time_axis) và W2(query_with_time_axis), sau đó áp dụng lớp Dense V.

- attention_weights = tf.nn.softmax(score, axis=1): Tính toán trọng số attention bằng cách áp dụng hàm softmax vào score theo chiều thứ 1.

- context_vector = attention_weights * values_with_time_axis: Tính toán context vector bằng cách nhân element-wise giữa attention_weights và values_with_time_axis.

- context_vector = tf.reduce_sum(context_vector, axis=1): Tính tổng của context_vector theo chiều thứ 1 để thu được context vector cuối cùng.

Như vậy, lớp BahdanauAttention giúp tính toán context vector và trọng số attention từ các query và values đầu vào. Điều này giúp mô hình học cách tập trung vào những thông tin quan trọng trong quá trình học và dự đoán.

```python
def load_and_combine_datasets(squad_csv, newqas_csv, val_squad_csv):
    squad_df = pd.read_csv(squad_csv)
    newqas_df = pd.read_csv(newqas_csv)
    val_squad_df = pd.read_csv(val_squad_csv)

    data = []
    for _, row in squad_df.iterrows():
        data.append({'question': row['question'], 'answer': row['text']})
    for _, row in newqas_df.iterrows():
        if isinstance(row['validated_answers'], str):
            for ans in row['validated_answers'].split('|'):
                data.append({'question': row['question'], 'answer': ans})
    return data, val_squad_df
```

Định nghĩa một hàm có tên là load_and_combine_datasets, nhận vào ba tham số: squad_csv, newqas_csv và val_squad_csv. Mục đích của hàm này là đọc ba tập dữ liệu từ các tệp CSV, kết hợp chúng và trả về dữ liệu đã kết hợp cùng với tập dữ liệu đánh giá. Dưới đây là một giải thích chi tiết hơn về từng dòng code trong hàm:

- newqas_df = pd.read_csv(newqas_csv): Đọc tệp CSV chứa dữ liệu NewQAs và lưu vào biến newqas_df dưới dạng DataFrame của pandas.

- val_squad_df = pd.read_csv(val_squad_csv): Đọc tệp CSV chứa dữ liệu đánh giá SQuAD và lưu vào biến val_squad_df dưới dạng DataFrame của pandas.

- data = []: Khởi tạo một danh sách trống tên là data để lưu trữ dữ liệu đã kết hợp từ squad_df và newqas_df.

- for _, row in squad_df.iterrows():: Duyệt qua từng dòng trong squad_df bằng cách sử dụng phương thức iterrows(). Biến row đại diện cho một dòng của DataFrame.

- data.append({'question': row['question'], 'answer': row['text']}): Thêm một từ điển vào danh sách data với hai khóa: 'question' và 'answer'. Giá trị của 'question' được lấy từ cột 'question' của row, và giá trị của 'answer' được lấy từ cột 'text' của row.

- for _, row in newqas_df.iterrows():: Tương tự như trên, duyệt qua từng dòng trong newqas_df bằng cách sử dụng phương thức iterrows().

- if isinstance(row['validated_answers'], str):: Kiểm tra xem giá trị của cột 'validated_answers' trong row có phải là một chuỗi hay không. Nếu có, tiếp tục thực hiện các dòng code trong khối lệnh này.

- for ans in row['validated_answers'].split('|'):: Tách chuỗi giá trị của cột 'validated_answers' thành một danh sách các câu trả lời bằng cách sử dụng phương thức split() với dấu phân tách '|'. Sau đó, duyệt qua từng câu trả lời ans trong danh sách.

- data.append({'question': row['question'], 'answer': ans}): Thêm một từ điển vào danh sách data với hai khóa: 'question' và 'answer'. Giá trị của 'question' được lấy từ cột 'question' của row, và giá trị của 'answer' được lấy từ biến ans (câu trả lời được tách từ chuỗi 'validated_answers').

- return data, val_squad_df: Sau khi đã kết hợp dữ liệu từ squad_df và newqas_df thành danh sách data, hàm trả về hai giá trị: danh sách data và DataFrame val_squad_df (dữ liệu đánh giá SQuAD).

```python
def extract_questions_and_answers(data):
    questions = [item['question'] for item in data]
    answers = [item['answer'] for item in data]
    return questions, answers
```

Định nghĩa một hàm có tên là extract_questions_and_answers, nhận vào một tham số data. data được giả định là một danh sách các từ điển, trong đó mỗi từ điển chứa hai khóa: 'question' và 'answer'. Mục đích của hàm này là trích xuất các câu hỏi và câu trả lời từ dữ liệu đầu vào và trả về chúng dưới dạng hai danh sách riêng biệt.


- questions = [item['question'] for item in data]: Dòng này sử dụng list comprehension (cấu trúc tạo danh sách ngắn gọn) để tạo một danh sách mới chỉ chứa các câu hỏi. Với mỗi phần tử item trong data, ta lấy giá trị của khóa 'question' và thêm nó vào danh sách questions.

- answers = [item['answer'] for item in data]: Tương tự như dòng trên, dòng này cũng sử dụng list comprehension để tạo một danh sách mới chỉ chứa các câu trả lời. Với mỗi phần tử item trong data, ta lấy giá trị của khóa 'answer' và thêm nó vào danh sách answers.

- return questions, answers: Cuối cùng, hàm trả về hai danh sách questions và answers dưới dạng một tuple. Danh sách questions chứa các câu hỏi và danh sách answers chứa các câu trả lời tương ứng.

```python
def tokenize_and_pad_sequences(questions, answers, max_len=100):
    tokenizer = Tokenizer()
    
    # Convert non-string values to empty strings
    questions = [str(q) if isinstance(q, str) else '' for q in questions]
    answers = [str(a) if isinstance(a, str) else '' for a in answers]
    
    tokenizer.fit_on_texts(questions + answers)
    sequences = tokenizer.texts_to_sequences(questions + answers)
    sequences_padded = pad_sequences(sequences, maxlen=max_len)
    
    question_seqs_padded = sequences_padded[:len(questions)]
    answer_seqs_padded = sequences_padded[len(questions):]
    
    return question_seqs_padded, answer_seqs_padded, tokenizer
```

Hàm tokenize_and_pad_sequences:

- def tokenize_and_pad_sequences(questions, answers, max_len=100): Hàm này nhận vào danh sách questions, answers và max_len (độ dài tối đa của chuỗi sau khi padding), mặc định là 100.

- tokenizer = Tokenizer(): Khởi tạo một đối tượng Tokenizer từ thư viện Keras.

- questions = [str(q) if isinstance(q, str) else '' for q in questions]: Chuyển đổi các giá trị không phải chuỗi trong questions thành chuỗi rỗng.

- answers = [str(a) if isinstance(a, str) else '' for a in answers]: Chuyển đổi các giá trị không phải chuỗi trong answers thành chuỗi rỗng.

- tokenizer.fit_on_texts(questions + answers): Huấn luyện tokenizer trên toàn bộ dữ liệu questions và answers.

- sequences = tokenizer.texts_to_sequences(questions + answers): Chuyển đổi các câu hỏi và câu trả lời sang dạng số bằng cách áp dụng tokenizer đã huấn luyện.

- sequences_padded = pad_sequences(sequences, maxlen=max_len): Padding các chuỗi số để có độ dài tối đa là max_len bằng cách thêm số 0 vào cuối mỗi chuỗi.

- question_seqs_padded = sequences_padded[:len(questions)]: Lấy phần đầu của sequences_padded tương ứng với số lượng câu hỏi, tạo thành question_seqs_padded.

- answer_seqs_padded = sequences_padded[len(questions):]: Lấy phần còn lại của sequences_padded tương ứng với số lượng câu trả lời, tạo thành answer_seqs_padded.

- return question_seqs_padded, answer_seqs_padded, tokenizer: Trả về question_seqs_padded, answer_seqs_padded và tokenizer.

Như vậy, hàm tokenize_and_pad_sequences thực hiện việc chuyển đổi các câu hỏi và câu trả lời sang dạng số và padding chúng để có cùng độ dài, chuẩn bị dữ liệu để đưa vào mô hình huấn luyện.

```python
def build_lstm_model(embedding_dim, max_len):
    model = tf.keras.Sequential([
        tf.keras.layers.Embedding(input_dim=embedding_dim, output_dim=max_len, input_length=max_len),
        tf.keras.layers.LSTM(64, return_sequences=False),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(1, activation='linear')
    ])
    return model

def build_bilstm_model(embedding_dim, max_len):
    model = tf.keras.Sequential([
        tf.keras.layers.Embedding(input_dim=embedding_dim, output_dim=max_len, input_length=max_len),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=False)),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(1, activation='linear')
    ])
    return model
```

Hàm mô hình LSTM:

Hàm này nhận vào hai tham số: embedding_dim (số lượng từ trong bộ từ vựng) và max_len (độ dài tối đa của chuỗi sau khi padding).

Khởi tạo một mô hình tuần tự (Sequential) gồm các lớp sau:

a. Embedding: Lớp nhúng để chuyển đổi các từ vựng sang dạng vector. Tham số input_dim là kích thước từ điển, output_dim là độ dài của chuỗi đầu ra, và input_length là độ dài của chuỗi đầu vào.

b. LSTM: Lớp LSTM với 64 đơn vị, chỉ trả về trạng thái ẩn cuối cùng (return_sequences=False).

c. Dense: Lớp kết nối đầy đủ với 32 đơn vị và hàm kích hoạt 'relu'.

d. Dense: Lớp kết nối đầy đủ với 1 đơn vị và hàm kích hoạt 'linear'.

Trả về mô hình đã xây dựng.

Hàm mô hình BiLSTM:

Hàm này cũng nhận vào hai tham số embedding_dim và max_len tương tự như hàm build_lstm_model.

Khởi tạo một mô hình tuần tự (Sequential) gồm các lớp sau:

a. Embedding: Tương tự như trong hàm build_lstm_model.

b. Bidirectional: Lớp này bao gồm một LSTM hai chiều với 64 đơn vị, chỉ trả về trạng thái ẩn cuối cùng (return_sequences=False). Mạng LSTM hai chiều sẽ học cả hai hướng từ trái sang phải và từ phải sang trái của dữ liệu đầu vào.

c. Dense: Lớp kết nối đầy đủ với 32 đơn vị và hàm kích hoạt 'relu'.

d. Dense: Lớp kết nối đầy đủ với 1 đơn vị và hàm kích hoạt 'linear'.

Trả về mô hình đã xây dựng.

```python
def build_lstm_with_attention_model(embedding_dim, max_len):
    inputs = tf.keras.Input(shape=(max_len,))
    x = tf.keras.layers.Embedding(input_dim=embedding_dim, output_dim=max_len, input_length=max_len)(inputs)
    x, state_h, state_c = tf.keras.layers.LSTM(64, return_sequences=True, return_state=True)(x)
    query_seq_encoding = tf.keras.layers.Concatenate()([state_h, state_c])
    attention = BahdanauAttention(128)
    context_vector, attention_weights = attention(query_seq_encoding, x)
    x = tf.keras.layers.Dense(32, activation='relu')(context_vector)
    x = tf.keras.layers.Dense(1, activation='linear')(x)
    model = tf.keras.Model(inputs=inputs, outputs=x)
    return model

def build_bilstm_with_attention_model(embedding_dim, max_len):
    inputs = tf.keras.Input(shape=(max_len,))
    x = tf.keras.layers.Embedding(input_dim=embedding_dim, output_dim=max_len, input_length=max_len)(inputs)
    x, forward_h, forward_c, backward_h, backward_c = tf.keras.layers.Bidirectional(
        tf.keras.layers.LSTM(64, return_sequences=True, return_state=True))(x)
    state_h = tf.keras.layers.Concatenate()([forward_h, backward_h])
    state_c = tf.keras.layers.Concatenate()([forward_c, backward_c])
    query_seq_encoding = tf.keras.layers.Concatenate()([state_h, state_c])
    attention = BahdanauAttention(128)
    context_vector, attention_weights = attention(query_seq_encoding, x)
    x = tf.keras.layers.Dense(32, activation='relu')(context_vector)
    x = tf.keras.layers.Dense(1, activation='linear')(x)
    model = tf.keras.Model(inputs=inputs, outputs=x)
    return model
```

Mô hình LSTM with attention: kết hợp với cơ chế Attention Bahdanau. Đây là một cách tiếp cận khá phổ biến để giải quyết các vấn đề liên quan đến chuỗi và ngôn ngữ. Dưới đây là giải thích từng bước của đoạn mã:

- Định nghĩa hàm build_lstm_with_attention_model(embedding_dim, max_len), nhận vào hai tham số: embedding_dim là kích thước từ điển của các từ và max_len là chiều dài tối đa của các chuỗi đầu vào.

- Tạo một đầu vào inputs với kích thước (max_len,).

- Thêm một lớp Embedding để chuyển đổi các từ thành vector đặc trưng phù hợp. Lớp này sẽ nhận đầu vào là inputs và có kích thước đầu ra là max_len.

- Thêm một lớp LSTM với 64 đơn vị, thiết lập return_sequences=True và return_state=True để trả về cả chuỗi ẩn và trạng thái ẩn cuối cùng. Lớp LSTM nhận đầu vào là kết quả của lớp Embedding.

- Nối hai trạng thái ẩn cuối cùng state_h và state_c để tạo ra một biểu diễn mã hóa chuỗi câu hỏi query_seq_encoding.

- Khởi tạo một đối tượng Attention (BahdanauAttention) với 128 đơn vị.

- Gọi phương thức Attention với query_seq_encoding và chuỗi ẩn x để tính context_vector và attention_weights.

- Thêm một lớp Dense với 32 đơn vị và hàm kích hoạt 'relu'. Lớp này nhận đầu vào là context_vector.

- Thêm một lớp Dense cuối cùng với 1 đơn vị và hàm kích hoạt 'linear' để tính giá trị đầu ra.

- Tạo một đối tượng Model với đầu vào là inputs và đầu ra là kết quả của lớp Dense cuối cùng. Trả về mô hình đã xây dựng.

- Định nghĩa hàm build_bilstm_with_attention_model(embedding_dim, max_len), nhận vào hai tham số: embedding_dim là kích thước từ điển của các từ và max_len là chiều dài tối đa của các chuỗi đầu vào.

- Trạng thái ẩn cuối cùng của LSTM theo cả hai hướng (forward và backward) được nối lại để tạo ra state_h và state_c.

- Nối hai trạng thái ẩn state_h và state_c để tạo ra một biểu diễn mã hóa chuỗi câu hỏi query_seq_encoding.

- Khởi tạo một đối tượng Attention (BahdanauAttention) với 128 đơn vị.

- Gọi phương thức Attention với query_seq_encoding và chuỗi ẩn x để tính context_vector và attention_weights.

- Thêm một lớp Dense với 32 đơn vị và hàm kích hoạt 'relu'. Lớp này nhận đầu vào là context_vector.

Kết quả là mô hình có thể đạt được độ chính xác cao hơn khi dự đoán hoặc tìm ra các câu trả lời phù hợp cho các câu hỏi được đưa ra.


```python
def train_and_evaluate_models(models, X_train, X_test, y_train, y_test, epochs=10, batch_size=32):
    results = []
    for model in models:
        model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
        history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.2)
        test_loss, test_acc = model.evaluate(X_test, y_test)
        results.append((history, test_loss, test_acc))
    return results
```

Định nghĩa một hàm train_and_evaluate_models được sử dụng để huấn luyện và đánh giá các mô hình trong danh sách models trên tập dữ liệu train (X_train, y_train) và tập dữ liệu test (X_test, y_test).

- Bên trong hàm:
